```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.datasets import make_regression
```

```python
data = make_regression(n_samples=15, noise=3, n_features=1)
```

```python
data
```

```python
inp = data[0]
out = data[1]
```

```python
plt.scatter(inp, out)
```

```python
model = LinearRegression()
```

```python
model.fit(inp, out)
```

```python
m = model.coef_
b = model.intercept_
```

```python
b
```

```python
m
```

```python
y_pred = model.predict(inp)
```

```python
plt.scatter(inp, out)
plt.plot(inp, y_pred, color="red")
```

```python
lr = 0.01
b = 1
m = 10
epochs = 50

for i in range(epochs):
    
    b_slope = -2 * np.sum(out - m * inp.ravel() - b)
    print(f"m={m}, b={b}, slope of b={b_slope}")
    m_slope = -2 * np.sum((out - m * inp.ravel() - b)* inp.ravel())
    print(f"m={m}, b={b}, slope of m={m_slope}")
    
    b = b - (lr * b_slope)
    print(f"updated b={b}")
    m = m - (lr * m_slope)
    print(f"updated m={m}")

#     temp = m*inp + b
#     plt.scatter(inp, out)
#     plt.plot(inp, y_pred, color="red")
#     plt.plot(inp, temp, color="blue")
# plt.show()
```
